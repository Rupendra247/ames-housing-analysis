```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Housing.csv')
print(df.to_string())
```

```python
val = pd.DataFrame(df)
print(val.to_string())
```

```python
print(df.drop_duplicates())
```

```python
df = df.replace({'yes': 1, 'no': 0})
print(df.to_string())
```

```python
df['area'].corr(df['price'])
```

```python
df['mainroad'].value_counts(normalize=True) * 100
sns.boxplot(x='mainroad', y='price', data=df)
plt.show()
```

```python
pd.crosstab(df['mainroad'], df['airconditioning'])
```

```python
x = df['price']
y = df['area']
plt.xlabel('Price')
plt.ylabel('area')
plt.title('Price and Area comnparison of House')
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('price')
plt.ylabel('area')
plt.ticklabel_format(style='plain',axis='both')
plt.scatter(x,y)
plt.show()
```

```python
sns.set_theme(style='whitegrid')
sns.histplot(df['price'], color='skyblue')
plt.xlabel('Price')
plt.ylabel('frequency')
plt.title('Price and Area comnparison of House')
plt.ticklabel_format(style='plain',axis='both')
```

```python
sns.boxplot(x='airconditioning', y='price', data=df, palette='Set1')
plt.title('Impact of Air Conditioning on House Price', fontsize=15)
plt.xlabel('Air Conditioning (yes/no)')
plt.ylabel('Price')
```

```python
df['mainroad'].value_counts().plot(kind='bar')
plt.title("Mainroad Distribution")
plt.show()
```

```python
df.groupby('bedrooms')['price'].mean().plot(kind='bar')
plt.title("Avg Price by Bedrooms")
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
df.boxplot(column='price')
plt.show()
```

```python
df['mainroad'] = df['mainroad'].map({'yes':1, 'no':0})
# sns.pairplot(df)
```

```python
print("Mean price:", df['price'].mean())
print("Median price:", df['price'].median())
print("Mode furnishing:", df['furnishingstatus'].mode())
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(df['price'], bins=30, color='skyblue')
plt.title(f"Price Distribution (skew={df['price'].skew():.2f})")

plt.subplot(1,2,2)
plt.hist(df['area'], bins=30, color='salmon')
plt.title(f"Area Distribution (skew={df['area'].skew():.2f})")

plt.show()
```

```python
plt.scatter(df['area'], df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Price vs Area")
plt.show()
```

```python
categorical_cols = ['guestroom','basement','hotwaterheating','airconditioning','furnishingstatus']

for col in categorical_cols:
    plt.figure()
    df[col].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
    plt.title(f"{col} Distribution")
    plt.ylabel("")
    plt.show()
```
